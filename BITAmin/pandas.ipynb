```python
import pandas as pd
```

1. 리스트 -> Series

```python
A = ['2023-01-29', '홍길동', 'ESFP', '0123']
pd.Series(A)
```

2. 딕셔너리 -> DataFrame

```python
B = {'사람1':[95, 100, 90, 80], '사람2':[85, 85, 90, 100], '사람3':[80, 90, 80, 90]}
df = pd.DataFrame.from_dict(data=B, orient='index', columns=['국어', '수학', '영어', '탐구'])
print(df)
```

3. 열(column) 삭제

```python
df.drop(['탐구'], axis=1, inplace=True)
print(df)
```

4. 열(column) 평균 + 반올림

```python
df['평균'] = round(df.mean(axis=1), 2)
print(df)
```

5. 등급 추가

```python
grades = [] #리스트 생성
for row in df['평균'] : #'average' 값을 얻어옴
    if row >= 95:
        grades.append('A+') #평균이 90점 이상이면 grades에 A를 추가
    elif row >= 90:
        grades.append('A')
    elif row >= 85:
        grades.append('B+')
    elif row >= 80 :
        grades.append('B')
        
df['grade'] = grades #grade열에 순차대로 원소를 대입

print(df)
```
